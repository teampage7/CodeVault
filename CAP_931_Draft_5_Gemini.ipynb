```python
# ============================================================================
# STEP 1: Install Dependencies
# ============================================================================

!pip install -q google-genai

print("Dependencies installed successfully!")

# ============================================================================
# STEP 2: Import Libraries
# ============================================================================

from google import genai
from google.genai import types
import json
from datetime import datetime

print("Libraries imported successfully!")

# ============================================================================
# STEP 3: API Key Setup
# ============================================================================

# Get API key from user
print("\nGet your FREE API key at: https://ai.google.dev")
print("No credit card required!\n")

api_key = input("Enter your Google Gemini API Key: ")

# Initialize the client
client = genai.Client(api_key=api_key)

print("API key configured successfully!")

# ============================================================================
# STEP 4: Define Agent Functions
# ============================================================================

def strategy_agent(company_name, product_name, category, known_info):
    """
    Agent 1: Analyzes company strategy and identifies opportunities
    """
    prompt = f"""You are a business strategy analyst. Analyze the following company for a sales opportunity:

Company: {company_name}
Product to sell: {product_name}
Product category: {category}
Known information: {known_info if known_info else 'None provided'}

Provide a strategic analysis covering:
1. Company's likely business priorities and challenges
2. How they might benefit from {category} solutions
3. Strategic initiatives they may be pursuing
4. Key business drivers and pain points

Keep response focused and under 300 words."""

    response = client.models.generate_content(
        model='gemini-2.0-flash-exp',
        contents=prompt
    )
    return response.text


def competitive_agent(company_name, product_name, competitors):
    """
    Agent 2: Analyzes competitive landscape
    """
    prompt = f"""You are a competitive intelligence analyst. Analyze the competitive landscape:

Company: {company_name}
Our product: {product_name}
Known competitors: {competitors if competitors else 'Research typical competitors in this space'}

Provide analysis of:
1. Competitive threats and alternatives the company may be considering
2. How our product differentiates from competitors
3. Potential objections based on competitive offerings
4. Unique selling points to emphasize

Keep response focused and under 300 words."""

    response = client.models.generate_content(
        model='gemini-2.0-flash-exp',
        contents=prompt
    )
    return response.text


def leadership_agent(company_name, contact_name, contact_title):
    """
    Agent 3: Analyzes leadership and decision-maker profile
    """
    prompt = f"""You are an executive profiling specialist. Analyze this decision-maker:

Company: {company_name}
Contact: {contact_name}
Title: {contact_title}

Provide insights on:
1. Typical priorities and concerns for someone in this role
2. Key performance metrics they likely care about
3. Decision-making authority and influence
4. Best approach for initial engagement

Keep response focused and under 250 words."""

    response = client.models.generate_content(
        model='gemini-2.0-flash-exp',
        contents=prompt
    )
    return response.text


def value_proposition_agent(company_name, product_name, value_prop, strategy_analysis, competitive_analysis):
    """
    Agent 4: Crafts tailored value proposition
    """
    prompt = f"""You are a value proposition specialist. Create a tailored pitch:

Company: {company_name}
Product: {product_name}
Core value proposition: {value_prop}

Strategy insights: {strategy_analysis}
Competitive insights: {competitive_analysis}

Create a compelling, personalized value proposition that:
1. Addresses their specific business challenges
2. Differentiates from competitors
3. Links to strategic priorities
4. Includes specific benefits and outcomes

Keep response focused and under 300 words."""

    response = client.models.generate_content(
        model='gemini-2.0-flash-exp',
        contents=prompt
    )
    return response.text


def synthesis_agent(company_name, contact_name, product_name,
                   strategy_analysis, competitive_analysis,
                   leadership_analysis, value_prop_analysis):
    """
    Agent 5: Synthesizes all insights into executive summary
    """
    prompt = f"""You are a sales strategist. Create an executive summary briefing:

Company: {company_name}
Contact: {contact_name}
Product: {product_name}

Strategy Analysis: {strategy_analysis}
Competitive Analysis: {competitive_analysis}
Leadership Analysis: {leadership_analysis}
Value Proposition: {value_prop_analysis}

Create a concise executive summary (under 200 words) that:
1. Highlights the top 3 reasons this is a good opportunity
2. Identifies the main value drivers
3. Suggests the best approach for initial outreach
4. Notes any key risks or concerns to address

Be specific and actionable."""

    response = client.models.generate_content(
        model='gemini-2.0-flash-exp',
        contents=prompt
    )
    return response.text

print("Agent functions defined successfully!")

# ============================================================================
# STEP 5: Collect User Inputs
# ============================================================================

print("\n" + "="*80)
print("SALES INTELLIGENCE BRIEFING GENERATOR")
print("="*80 + "\n")

# Collect inputs
product_name = input("Product Name: ")
product_category = input("Product Category (e.g., Cloud Data Platform): ")
value_proposition = input("Value Proposition (brief): ")
company_name = input("Target Company Name: ")
contact_name = input("Target Contact Name: ")
contact_title = input("Target Contact Title (e.g., Chief Data Officer): ")
competitors = input("Competitors (comma-separated, or press Enter to skip): ")
known_info = input("Any known company information (or press Enter to skip): ")

print("\nInputs collected successfully!")

# ============================================================================
# STEP 6: Run Multi-Agent Analysis
# ============================================================================

print("\n" + "="*80)
print("GENERATING INTELLIGENCE REPORT...")
print("="*80 + "\n")

try:
    # Initialize variables with defaults in case of errors
    strategy_analysis = "Analysis unavailable - Error occurred"
    competitive_analysis = "Analysis unavailable - Error occurred"
    leadership_analysis = "Analysis unavailable - Error occurred"
    value_prop_analysis = "Analysis unavailable - Error occurred"
    executive_summary = "Analysis unavailable - Error occurred"

    # Agent 1: Strategy Analysis
    print("Analyzing company strategy...")
    try:
        strategy_analysis = strategy_agent(company_name, product_name, product_category, known_info)
        print("Strategy analysis complete\n")
    except Exception as e:
        print(f"Strategy analysis failed: {str(e)}\n")

    # Agent 2: Competitive Analysis
    print("Analyzing competitive landscape...")
    try:
        competitive_analysis = competitive_agent(company_name, product_name, competitors)
        print("Competitor analysis complete\n")
    except Exception as e:
        print(f"Competitive analysis failed: {str(e)}\n")

    # Agent 3: Leadership Analysis
    print("Analyzing leadership...")
    try:
        leadership_analysis = leadership_agent(company_name, contact_name, contact_title)
        print("Leadership analysis complete\n")
    except Exception as e:
        print(f"Leadership analysis failed: {str(e)}\n")

    # Agent 4: Value Proposition
    print("Crafting value proposition...")
    try:
        value_prop_analysis = value_proposition_agent(
            company_name, product_name, value_proposition,
            strategy_analysis, competitive_analysis
        )
        print("Value proposition complete\n")
    except Exception as e:
        print(f"Value proposition failed: {str(e)}\n")

    # Agent 5: Executive Summary
    print("Creating executive summary...")
    try:
        executive_summary = synthesis_agent(
            company_name, contact_name, product_name,
            strategy_analysis, competitive_analysis,
            leadership_analysis, value_prop_analysis
        )
        print("Synthesis complete\n")
    except Exception as e:
        print(f"Executive summary failed: {str(e)}\n")

    print("\n" + "="*80)
    print("REPORT GENERATION COMPLETED")
    print("="*80 + "\n")

except Exception as e:
    print(f"\nCritical error: {str(e)}")
    print("Please check your API key and internet connection.")

# ============================================================================
# STEP 7: Format and Display Report
# ============================================================================

# Create timestamp
timestamp = datetime.now().strftime("%B %d, %Y at %I:%M %p")
file_timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

# Format company name for filename
company_file = company_name.replace(" ", "_").replace(".", "")

# Build report
report = f"""
{'='*80}
                    SALES INTELLIGENCE BRIEFING
{'='*80}
Generated: {timestamp}

{'-'*80}
PROSPECT OVERVIEW
{'-'*80}
Company:          {company_name}
Target Contact:   {contact_name} ({contact_title})
Product:          {product_name}
Category:         {product_category}
Competitors:      {competitors if competitors else 'To be researched'}

{'='*80}
EXECUTIVE SUMMARY
{'='*80}
{executive_summary}

{'='*80}
COMPANY STRATEGY ANALYSIS
{'='*80}
{strategy_analysis}

{'='*80}
COMPETITIVE LANDSCAPE
{'='*80}
{competitive_analysis}

{'='*80}
KEY LEADERSHIP INSIGHTS
{'='*80}
{leadership_analysis}

{'='*80}
TAILORED VALUE PROPOSITION
{'='*80}
{value_prop_analysis}

{'='*80}
RECOMMENDED TALKING POINTS
{'='*80}
1. STRATEGIC ALIGNMENT
   Reference their {product_category} initiatives and strategic direction

2. COMPETITIVE DIFFERENTIATION
   Position {product_name} against key competitors

3. DECISION MAKER ENGAGEMENT
   Target {contact_name} with insights above

{'='*80}
NEXT STEPS
{'='*80}
[ ] Review this briefing thoroughly before prospect outreach
[ ] Customize email/call script using specific insights above
[ ] Research any recent news or press releases
[ ] Prepare specific use cases relevant to their strategy
[ ] Schedule discovery call with {contact_name}

{'='*80}
DISCLAIMER
{'='*80}
This briefing was generated using AI-powered analysis. Verify all key details
and claims before customer engagement. Use as a starting point for research.

{'='*80}
"""

# Display report
print(report)

# ============================================================================
# STEP 8: Save Output Files
# ============================================================================

# Save text report
txt_filename = f"sales_briefing_{company_file}_{file_timestamp}.txt"
with open(txt_filename, 'w') as f:
    f.write(report)

print(f"\nReport saved as: {txt_filename}")
print("You can download it from the Colab files panel (left sidebar)")

# Save JSON data for integration
json_data = {
    "metadata": {
        "generated": timestamp,
        "company": company_name,
        "contact": contact_name,
        "product": product_name
    },
    "analysis": {
        "executive_summary": executive_summary,
        "strategy": strategy_analysis,
        "competitive": competitive_analysis,
        "leadership": leadership_analysis,
        "value_proposition": value_prop_analysis
    },
    "inputs": {
        "product_name": product_name,
        "product_category": product_category,
        "value_proposition": value_proposition,
        "company_name": company_name,
        "contact_name": contact_name,
        "contact_title": contact_title,
        "competitors": competitors,
        "known_info": known_info
    }
}

json_filename = f"sales_data_{company_file}_{file_timestamp}.json"
with open(json_filename, 'w') as f:
    json.dump(json_data, f, indent=2)

print(f"Data saved as: {json_filename}")

# ============================================================================
# COMPLETION
# ============================================================================

print("\n" + "="*80)
print("THANK YOU FOR USING OUR SERVICE!")
print("POWERED BY ANTONEO PAGE - AI PROMPT ENGINEERING - PER SCHOLAS")
print("="*80)
```
